Linear Reg, Coefficient, MAE, R2_Score, MSE, MAPE, train_test_split



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_squared_error, r2_score, mean_absolute_error
```

#Linear Regression on bmw dataset

```python
df = pd.read_csv('bmw.csv')
df.head()
```

```python
df.columns
```

```python
X = df.drop(columns = ['transmission','model', 'price', 'mileage', 'fuelType'], axis = 1)
Y = df['price']
```

# Train Test Split Explained

* **X_train:** feature columns for training
* **X_test:** feature columns for testing
* **Y_train:** target variable columns for training
* **Y_test:** target variable columns for testing   

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .20, random_state = 42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

#Building the Model



```python
model = LinearRegression()

model = model.fit(X_train, y_train)

# take the features (X_train --> col1, col3, col3...) and here is the actual price (y_train --> one single column [col])
# find the relation between them
```

```python
co_ef = model.coef_
print(co_ef)
```

```python
intercept = model.intercept_
print(intercept)

#y(x) = m1x1 + m2x2 + m3x3 +c
```

```python
f = X.columns
```

```python
y_x = pd.DataFrame({'frature': f,
                     'co_ef': np.round(co_ef,2)})
y_x.head()
```

```python
# now predict the price of the unseen data in X_test
y_pred = model.predict(X_test)
print(y_pred)
```

```python
lear = pd.DataFrame()
lear['price'] = y_test
lear['pred'] = y_pred
lear
```

```python
mse = mean_squared_error(y_test, y_pred)
print("mse = ", mse)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print("rmse = ", rmse)

mae = mean_absolute_error(y_test, y_pred)
print("mae = ", mae)


r2 = r2_score(y_test, y_pred)
print("r_squared = ", r2)
```

#Linear Regression on VW dataset

```python
data = pd.read_csv('vw.csv')
data
```

```python
print('--------------------------------------------------')
print('Unique Modles')
print('--------------------------------------------------')
display(data['model'].value_counts())

print('--------------------------------------------------')
print('Data Description')
print('--------------------------------------------------')
display(data.describe())

print('--------------------------------------------------')
print('Data Info')
print('--------------------------------------------------')
display(data.info())
```

```python
X = data.drop(columns= ['model', 'price', 'transmission', 'fuelType'], axis = 1)
Y = data['price']
display(X)
display(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    Y, 
                                                    test_size=.20, 
                                                    random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape) 
```

```python
model = LinearRegression()
model = model.fit(X_train, y_train)
X.columns
```

```python
co_ef = model.coef_
print("Coefficeint: ",co_ef)

int_cept = model.intercept_
print("Intercept: ",int_cept)

coef_df = pd.DataFrame({'feature_var': X.columns,
                        'Coefficient': co_ef})
display(coef_df)
```

```python
y_pred = model.predict(X_test)
pref_df = coef_df = pd.DataFrame({'actual_val': y_test,
                                'prediction': np.round(y_pred,2)})

pref_df.head(10)
```
